#Function to find the largest of three numbers.


```python
#Largest of 3 numbers
def largest_number(*args):
  print(args)
  max_num=max(args)
  return max_num
print(f"---------Largest of 3 numbers------------")
inp_num=tuple(map(int,input("Enter the numbers separated by , : ").split(',')))
print(f"Input Numbers = {inp_num}\nLargest of the input numbers = {largest_number(*inp_num)}")
```

#Factorial of a number using recursion.


```python
#Factorial of a number using recursion.
def factorial(num):
  if num==0 or num==1:
    return 1
  else:
    return num*factorial(num-1)
print(f"----------Factorial of a number-------------")
n=int(input("Enter the number to find the factorial : "))
print(f"Factorial of {n} = {factorial(n)}")
```

#String is a palindrome or not.

```python
#String is a palindrome or not.
def check_palindrome(inp_str):
  if inp_str[0:]==inp_str[::-1]:
    return f"{inp_str} is a palindrome"
  else:
    return f"{inp_str} is not a palindrome"
inp=input("Enter the string: ")
print(f"{check_palindrome(inp)}")
```

#Function using *args to find the average of multiple numbers.


```python
#Function using *args to find the average of multiple numbers.
def avg_numbers(*args):
  return sum(args)/len(args)
inp_num=tuple(map(int,input("Enter the numbers separated by ,: ").split(',')))
print(f"---------------Average of input list of numbers---------------\nInput List = {inp_num}")
print(f"Average of all the numbers = {avg_numbers(*inp_num):.2f}")
```

#Accepts a list of student marks and uses a loop to count how many students scored above 75.


```python
#Accepts a list of student marks and uses a loop to count how many students scored above 75.
def marks_analysis(*args):
  high_score={f"student {index}":mark for index,mark in enumerate(args,start=1) if mark>75 }
  return high_score
inp_marks=[100,56,89,67,99,56,92,49,87,76]
print(f"----------------Student Marks Analysis----------------------\nInput Marks = {inp_marks}")
print(f"\nStudents Scored above 75 are :\n{marks_analysis(*inp_marks)}")
```

#Takes a list of sales values and uses a loop to calculate the total sales greater than a given threshold.


```python
#Takes a list of sales values and uses a loop to calculate the total sales greater than a given threshold.
def sales_analysis(threshold,*args):
  high_sales=[sale for sale in args if sale>threshold]
  total_sales=sum(high_sales)
  return high_sales,total_sales
inp_sales=[1000,999,599,399,200,508,789,897,900,899,970]
inp_threshold=600
print(f"-------------------Sales Analysis----------------------\nInput Sale Value = {inp_sales}\nThreshold Set = {inp_threshold}")
high_sales,total_sales=sales_analysis(inp_threshold,*inp_sales)
print(f"\nSale values more than the given threshold({inp_threshold}) = {high_sales}\nTotal Sales more than the given threshold({inp_threshold}) = {total_sales}")
```

#Count the number of vowels present in a string


```python
#Count the number of vowels present in a string
def vowel_count(inp_str):
  vowel_list=[char for char in inp_str if char.isalpha() and char in 'aeiouAEIOU']
  return vowel_list
print(f"------------------Vowels Count---------------")
inp_str=input("Enter the srting to count the vowels : ")
out_list=vowel_count(inp_str)
print(f"\nVowel list in the '{inp_str}' = {out_list}\nNo of vowels present = {len(out_list)}")
```

#Takes a list of numbers and uses a loop to return only the even numbers in a new list.


```python
#Takes a list of numbers and uses a loop to return only the even numbers in a new list.
def even_numbers(inp_list):
  even_list=[val for val in inp_list if val%2==0]
  return even_list
inp_list=[100,2,3,9,89,78,90,31,67,86,24]
print(f"------------------------Even Numbers List------------------------\nInput List = {inp_list}")
out_list=even_numbers(inp_list)
print(f"\nEven Numbers List = {out_list}\nCount of even numbers = {len(out_list)}")
```

#Monthly Bonus for Employees
Write a function that accepts a list of employee sales.

If sales are greater than 50,000, bonus is 10% of sales
If sales are between 30,000 and 50,000, bonus is 5%
Otherwise, no bonus

```python
#Monthly Bonus for Employees
#Write a function that accepts a list of employee sales.
#If sales are greater than 50,000, bonus is 10% of sales
#If sales are between 30,000 and 50,000, bonus is 5%
#Otherwise, no bonus
def bonus_calculator(*args):
  bonus_dict={
              f"Employee {index}":sale*0.1 if sale>50000 else
              sale*0.05 if sale >=30000 and sale <50000 else
              0 for index,sale in enumerate(args,start=1)
              }
  return bonus_dict
inp_sales=[90000,50900,89100,34999,19000,23800,40900,78000,12000]
print(f"----------------Employee Bonus Calculator----------------\nInput Employee Sale Details = {inp_sales}")
bonus=bonus_calculator(*inp_sales)
print(f"\nBonus Details :\n{bonus}")
```
